```python
mer_2010_set = set(mer_2010_gdf['CVEGEO'])
```

```python
mer_2020_set = set(mer_2020_gdf['CVEGEO'])
```

```python
remains_2010_to_2020 = mer_2010_set & mer_2020_set
```

```python
new_in_2020 = mer_2020_set - mer_2010_set
```

```python
disappears_2010_to_2020 = mer_2010_set - mer_2020_set
```

Using sets to create specific geodataframes

```python
remains_2010_to_2020_gdf = mer_2020_gdf[mer_2020_gdf["CVEGEO"].isin(remains_2010_to_2020)]
```

```python
new_in_2020_gdf = mer_2020_gdf[mer_2020_gdf["CVEGEO"].isin(new_in_2020)]
```

```python
disappears_2010_to_2020_gdf = mer_2010_gdf[mer_2010_gdf['CVEGEO'].isin(disappears_2010_to_2020)]
```

Convert to general crs

```python
print(remains_2010_to_2020_gdf.crs)
```

```python
print(new_in_2020_gdf.crs)
```

```python
print(disappears_2010_to_2020_gdf.crs)
```

```python
remains_2010_to_2020_gdf = remains_2010_to_2020_gdf.to_crs("EPSG:4326")
new_in_2020_gdf = new_in_2020_gdf.to_crs("EPSG:4326")
disappears_2010_to_2020_gdf = disappears_2010_to_2020_gdf.to_crs("EPSG:4326")
```

```python
remains_2010_to_2020_json = remains_2010_to_2020_gdf.to_json()
new_in_2020_json = new_in_2020_gdf.to_json()
disappears_2010_to_2020_json = disappears_2010_to_2020_gdf.to_json()
```

Visualization

```python
import folium
```

```python
m1 = folium.Map(location=[20.9674, -89.5926], zoom_start=10)
m2 = folium.Map(location=[20.9674, -89.5926], zoom_start=10)
m3 = folium.Map(location=[20.9674, -89.5926], zoom_start=10)
```

```python
folium.GeoJson(remains_2010_to_2020_json, name="Remains 2010 to 2020").add_to(m1)
folium.GeoJson(new_in_2020_json, name="New in 2020").add_to(m2)
folium.GeoJson(disappears_2010_to_2020_json, name="Disappears 2010 to 2020").add_to(m3)
```
